# Pinwheel Tiling

```python
import time
from pinwheel import *
```

Start with the prototile:

```python
tiling = pinwheel()
tiling.patch
```

Inflate and subdivide:

```python
# run this cell to do a subsitution!
tiling.sub()
tiling.patch
```

Save the SVG:

```python
path = 'patch_'+time.strftime("%Y%m%d-%H%M%S")+'.svg'
tiling.save(path)
```
